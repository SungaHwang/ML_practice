# SVD

### 특이값 분해

* 고유값 분해와 다르게 정방행렬뿐만 아니라 행과 열의 크기가 다른 행렬에도 적용 가능 (PCA경우, 정방행렬만 고유벡터로 분해 가능)
* 직교하는 벡터 집합에 대하여(v) 선형 변환 후에 그 크기(시그마)는 변하지만, 여전히 직교할 수 있게 만드는 그 직교 벡터 집합은 무엇이고, 변형 후(U)의 결과는 무엇인가?

* U, A: 특이벡터 (singular vector)
* 시그마: 대각행렬- 특이값 (singular value)

### Truncated SVD

* 대각원소 중 상위 몇 개만 추출해서 여기에 대응하는 U와 V원소도 함게 제거해 차원 축소된 형태로 분해

```python
import numpy as np
from numpy.linalg import svd
```

```python
np.random.seed(121)
a = np.random.randn(4, 4)
print(np.round(a, 3))
```

```python
U, Sigma, Vt = svd(a)
print(U.shape, Sigma.shape, Vt.shape)
print('U matrix:\n', np.round(U, 3))
print('Sigma Value:\n', np.round(Sigma, 3))
print('V transpose matrix:\n', np.round(Vt, 3))
```

### 데이터가 컬럼 간 의존성이 있을 때

```python
# 3번째 = 1번째 + 2번째
# 4번째 = 1번쨰
a[2] = a[0] + a[1]
a[3] = a[0]
print(np.round(a, 3))
```

```python
U, Sigma, Vt = svd(a)
print(U.shape, Sigma.shape, Vt.shape)
print('Sigma Value:\n', np.round(Sigma, 3))
```

* Sigma값 중 2개가 0으로 변함-> 선형 독립인 컬럼 벡터의 개수가 2개
* U행렬 중 앞 두 열만 추출, Vt에서 앞 두 행만 추출

### SVD 적용

```python
from scipy.sparse.linalg import svds
from scipy.linalg import svd
```

```python
np.random.seed(121)
matrix = np.random.randn(6, 6)
print('원본 행렬:\n', matrix)

U, Sigma, Vt = svd(matrix, full_matrices = False)
print(U.shape, Sigma.shape, Vt.shape)
print('\n분해 행렬 차원:', U.shape, Sigma.shape, Vt.shape)
print('\nSigma값 행렬:', Sigma)
```

### Truncated SVD 적용

```python
num_components = 4

U_tr, Sigma_tr, Vt_tr = svds(matrix, k= num_components)
print('\nTruncated SVD 분해 행렬 차원:', U_tr.shape, Sigma_tr.shape, Vt_tr.shape)
print('\nTruncated SVD Sigma값 행렬:', Sigma_tr)
```

```python
num_components = 3

U_tr, Sigma_tr, Vt_tr = svds(matrix, k= num_components)
print('\nTruncated SVD 분해 행렬 차원:', U_tr.shape, Sigma_tr.shape, Vt_tr.shape)
print('\nTruncated SVD Sigma값 행렬:', Sigma_tr)
```

### 실습: iris data

```python
from sklearn.decomposition import TruncatedSVD, PCA
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = load_iris()
iris_ftrs = iris.data

tsvd = TruncatedSVD(n_components = 2)
tsvd.fit(iris_ftrs)
iris_tsvd = tsvd.transform(iris_ftrs)
```

```python
plt.scatter(x = iris_tsvd[:, 0], y = iris_tsvd[:, 1], c = iris.target)
plt.xlabel('TruncatedSVD Component 1')
plt.ylabel('TruncatedSVD Component 2')
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
iris_scaled = scaler.fit_transform(iris_ftrs)
```

```python
# truncated SVD
tsvd = TruncatedSVD(n_components = 2)
tsvd.fit(iris_scaled)
iris_tsvd = tsvd.transform(iris_scaled)
```

```python
# pca
pca = PCA(n_components =2)
pca.fit(iris_scaled)
iris_pca = pca.transform(iris_scaled)
```

```python
fig, (ax1, ax2) = plt.subplots(figsize = (9, 4), ncols = 2)
ax1.scatter(x = iris_tsvd[:, 0], y = iris_tsvd[:, 1], c= iris.target)
ax2.scatter(x = iris_pca[:, 0], y = iris_pca[:, 1], c = iris.target)
ax1.set_title('Truncated SVD Transformed')
ax2.set_title('PCA Transformed')
```
