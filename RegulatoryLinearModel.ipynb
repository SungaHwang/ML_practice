# 규제 선형 모델

### 규제: 비용 함수에 alpha 값으로 패널티를 부여해 회귀 계수 값의 크기를 감소시켜 과적합을 개선하는 방식

### 크게 L2방식과 L1 방식으로 구분
L2: W제곱에 패널티 부여 (릿지회귀)
, L1: W의 절대값에 패널티 부여 (라쏘회귀)

-좋은 머신러닝 회귀 모델이 되려면 선형 모델의 비용함수는 RSS가 최소화 되어야 함.

* RSS (비용cost)

하지만, RSS를 최소화 (실제 값과 예측값의 차이를 최소화하는 것만 고려) -> 과적합, 회귀 계수 크게 증가 -> 오히려 변동성 심해짐


** 따라서, 최적 모델을 위한 cost 함수 구성요소 = (학습데이터 잔차 오류 최소화 + 회귀계수 크기 제어) [균형]

회귀 계수의 크기를 제어해 과적합을 개선하기 위해 위의 식을 최소화 하는 것으로 변경

* alpha: 학습 데이터 적합 정도와 회귀 계수 값의 크기 제어를 수행하는 튜닝 파라미터
* W: 회귀 계수

* alpha= 0 인 경우는 W가 커도 alpha*||W||가 0이 되어 비용 함수는 Min(RSS(W))

* alpha = 무한대인 경우 alpha*||W||도 무한대가 되므로 비용 함수는 W를 0에 가깝게 최소화 해야 함

* alpha를 0에서부터 값을 증가시키면 회귀 계수 값의 크기 감소
